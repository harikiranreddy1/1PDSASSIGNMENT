```python
import matplotlib.pyplot as plt    
import seaborn as sb              
import pandas as pd
import warnings
warnings.filterwarnings('ignore')


clean_student_data = pd.read_csv('StudentsPerformance.csv')

clean_student_data['total score'] = clean_student_data['math score'] + clean_student_data['reading score'] + clean_student_data['writing score']
clean_student_data['average score'] = clean_student_data['total score'] / 3
clean_student_data.head()
#displaying dataframe
clean_student_data
```

```python
plt.rcParams['figure.figsize'] = (30, 12)

# Create a subplot with bar charts
plt.subplot(1, 4, 1)
ax =sb.histplot(clean_student_data, x = 'gender', color='black')
ax
```

```python
# Continue with similar code blocks for other pie charts
plt.subplot(1, 4, 2)
ax =sb.histplot(clean_student_data, x = 'race/ethnicity', color='red')
ax
```

```python
plt.subplot(1, 4, 3)
ax =sb.histplot(clean_student_data, x = 'parental level of education',color='blue')
plt.setp(ax.get_xticklabels(), rotation=30, horizontalalignment='right')
ax
```

```python
plt.subplot(1, 4, 4)
ax =sb.histplot(clean_student_data, x = 'lunch', color='green')
ax
```

```python
1) From above we can interpret that, data is not biased with gender as data has almost same number of samples for male and female.
2) Around 30% students are of ethnicity group C in this data sample are  4 times larger than group A.
3) Students whose parents have a bachelor's or master's degree together are less than those in the other three scenarios.
4)Standard lunches are higher than free or reduced lunches for students.
```

```python

sb.barplot(data=clean_student_data,x='race/ethnicity',y='math score',hue='gender')
```

```python
The graphic above compares gender-based average maths score to race/ethnicity. We can infer from the preceding graph that men have much higher averages than women, and that this difference is independent of ethnicity. however, compared to all other ethnic groups, group E males and girls have higher math averages.
```

```python
plt.subplots(1,3,figsize=(20,20))

plt.subplot(131)
ax =sb.histplot(data=clean_student_data,x='average score',hue='lunch')

plt.subplot(132)
ax =sb.histplot(data=clean_student_data[clean_student_data.gender=='male'],x='average score',hue='lunch')

plt.subplot(133)
ax =sb.histplot(data=clean_student_data[clean_student_data.gender=='female'],x='average score',hue='lunch')
plt.show()

```

```python
Around 350 students are served free/reduced.rest all are standard lunch. Now, from above histogram we can interpret that, overall having standard lunch has higher average score than free/reduced lunch irrespective of gender.
```

```python
sb.violinplot(clean_student_data,x="parental level of education",y="average score",hue='gender', split=True)
```

```python
From the above figure we can say that males whose parents have masters degree perform averagely slightly better than females. companred to other qualifications. and in general females perform better than males irrespective of their parents education. students whose parents are master degree holders perform well overall. but inverse need not be true.
```

```python
sb.catplot(data=clean_student_data,x='gender',y='reading score',kind='box',col='test preparation course', whis=[0,100])
plt.show()
```

```python
from the above figure we can say, irrepesective of gender, students who have completed course scored higher averages in reading than students who dont have completed course/not taken any course. and in general female students scored more in reading than male students. this is completely opposite to math scores.
```
